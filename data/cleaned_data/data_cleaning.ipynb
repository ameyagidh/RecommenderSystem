```python
# Read in csv file:
import pandas as pd
pd.set_option('display.max_columns', None)
df_course = pd.read_csv('newcourses.csv')
df_student = pd.read_csv('newstudents.csv')
df_plan = pd.read_csv('newplans.csv')
```

```python
df_plan.head()
```

```python
df_id = pd.DataFrame(df_plan, columns=['STUDENT_ANONYMOUS_ID'])
df_id.describe()
```

```python
df_id['new_id'] = range(100000, 100000+len(df_id))
df_id.head()
```

```python
new_df_plan = df_plan.join(df_id.set_index('STUDENT_ANONYMOUS_ID'),on='STUDENT_ANONYMOUS_ID')
new_df_plan.head()
```

```python
new_id_df_plan = new_df_plan.drop(['STUDENT_ANONYMOUS_ID'], axis=1)
new_id_df_plan.head()
```

```python
clean_df_plan = new_id_df_plan.drop(['MOST_RECENT_TERM','FIRST_SCHOOL','FIRST_PLAN_TYPE','GRADUATED_FLAG','GRADUATION_DATE'], axis=1)
clean_df_plan.head()
```

```python
clean_df_plan.FIRST_PLAN.unique()
```

```python
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Information Science'] = 1
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Library and Information Sci'] = 2
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Library & Information Science'] = 2
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Telecommunications'] = 3
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Computer Science'] = 4
clean_df_plan.FIRST_PLAN[clean_df_plan.FIRST_PLAN == 'Intelligent Systems'] = 5
clean_df_plan.head()
```

```python
df_degree = pd.read_csv('degree.csv')
df_degree.head()
```

```python
clean_df_plan.FIRST_DEGREE_TYPE.unique()
```

```python
clean_df_plan.FIRST_DEGREE_TYPE[clean_df_plan.FIRST_DEGREE_TYPE == 'Baccalaureate Degree'] = 1
clean_df_plan.FIRST_DEGREE_TYPE[clean_df_plan.FIRST_DEGREE_TYPE == 'Masters Degree'] = 2
clean_df_plan.FIRST_DEGREE_TYPE[clean_df_plan.FIRST_DEGREE_TYPE == 'Doctorate Degree'] = 3
clean_df_plan.head()
```

```python
clean_df_plan['email'] = ''
clean_df_plan.head()
```

```python
clean_df_plan['name'] = ''
clean_df_plan['username'] = ''
clean_df_plan['password'] = ''
clean_df_plan['age'] = ''
clean_df_plan['gender'] = ''
clean_df_plan['active'] = 'N'
clean_df_plan.head()
```

```python
clean_df_plan = clean_df_plan.rename(columns={'CAREER_LEVEL':'level_id','START_TERM':'start_term','FIRST_PLAN':'plan_id','FIRST_DEGREE_TYPE':'degree_id','new_id':'student_id'})
clean_df_plan.head()
```

```python
clean_df_plan.to_csv("student_c.csv",index=False)
```

```python
new_df_course.ACADEMIC_CAREER_CD[new_df_course.ACADEMIC_CAREER_CD == 'UGRD'] = 'UG'
new_df_course.ACADEMIC_CAREER_CD[new_df_course.ACADEMIC_CAREER_CD == 'GRAD'] = 'GR'
new_df_course.SUBJECT_DESCR[new_df_course.SUBJECT_DESCR == 'Computer Science'] = 4
new_df_course.SUBJECT_DESCR[new_df_course.SUBJECT_DESCR == 'Information Science'] = 1
new_df_course.head()
```

```python
new_df_course.to_csv("course_c.csv",index=False)
```
